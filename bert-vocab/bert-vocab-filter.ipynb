## Read the vocab.text file for BERT

```python
with open("BERT-vocab.txt", encoding="utf8") as txt:
    f = txt.readlines()
```

```python
import enchant
import re
```

### The following preprocessing tasks are necessary
1. Remove \n new line characters
2. Remove words enclosed in [..]
3. Remove single entry letters
4. Remove numeric characters


```python
word_list = f

# 1. Replace new line '\n' with a null character
word_list = [word.replace('\n', '') for word in word_list]

# 2. Remove words enclosed in [..]
word_list = [word for word in word_list if not re.match(r'\[.*\]', word)]

# 3. Remove words starting with special characters and alphanumeric characters
# Define a regular expression pattern to match words with only letters
pattern = re.compile("^[A-Za-z]+$")

# Filter out words that match the pattern
word_list = [word for word in word_list if pattern.match(word)]

# 4. Remove single entry letters
word_list = [word for word in word_list if len(word) > 1]
```

```python
import tqdm
li = []
d = enchant.Dict("en_US")
for i in tqdm.tqdm(word_list):
    word = i.replace("\n", "")
    if d.check(word):
        li.append(word)
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")
```

```python
stemmed_words = []
for word in li:
    stemmed_words.append(stemmer.stem(word))

stemmed_words_set = sorted(list(set(stemmed_words)))
```
